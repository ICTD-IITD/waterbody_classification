```python
import pandas as pd

#In CSV rename system:index to system_index and .geo to geo

pd.set_option('display.max_rows', 50)


df = pd.read_csv('GroundTruth/seasonal_6_tile.csv')
```

```python
df.rename(columns = {'system:index':'system_index'}, inplace = True)
df.rename(columns = {'.geo':'geo'}, inplace = True)
```

```python
dfn = df
dfn["system_index"] = df["system_index"].str[:6]

dfn.fillna(0) 
```

```python
df = df[df.Name.str.startswith('seasonal_6_tile')]
```

```python
print(df)
```

```python
new_df = df[['geo']].drop_duplicates()
```

```python
len(new_df)
```

```python
# "Kharif"
new_df['11'] = 0
new_df['12'] = 0
new_df['13'] = 0
new_df['14'] = 0

# "Rabi"
new_df['21'] = 0
new_df['22'] = 0
new_df['23'] = 0
new_df['24'] = 0

# "Zaid"
new_df['31'] = 0
new_df['32'] = 0
new_df['33'] = 0
new_df['34'] = 0
```

```python
new_df.head()
```

```python
for index, row in df.iterrows():
    month = (int)(row['system_index'][4:6])
    classif = row['classification']
    # print(classif)
    
    if month >= 7 and month <= 10:
        t = '1'
    elif month >= 3 and month <= 6:
        t = '3'
    else:
        t = '2'
    
    column_name = t + str(classif)
    # print(column_name)
    
    new_df.loc[new_df['geo'] == row['geo'], [column_name]] += 1
    # dfn.loc[dfn['geo'] == some_value]
    
```

```python
print(new_df)
```

### COLOR CODING
- 0 => No water
- 1 => Kharif
- 2 => Rabi
- 3 => Zaid
- 4 => Kharif and Rabi
- 5 => Kharif and Zaid
- 6 => Zaid and Rabi
- 7 => Kharif, Rabi, and Zaid 

```python
new_df['CODE'] = new_df['11']
```

```python
for i in range(len(new_df)):
  label = 0
  in_kharif = False
  in_zaid = False
  in_rabi = False
  if new_df['11'][i] > 0 or new_df['12'][i] > 0 or new_df['13'][i] > 0 or new_df['14'][i] > 0:
    in_kharif = True
  if new_df['21'][i] > 0 or new_df['22'][i] > 0 or new_df['23'][i] > 0 or new_df['24'][i] > 0:
    in_rabi = True
  if new_df['31'][i] > 0 or new_df['32'][i] > 0 or new_df['33'][i] > 0 or new_df['34'][i] > 0:
    in_zaid = True

  if in_kharif and in_rabi and in_zaid:
    new_df['CODE'] = 7
  if not(in_kharif) and in_rabi and in_zaid:
    new_df['CODE'] = 6
  if in_kharif and not(in_rabi) and in_zaid:
    new_df['CODE'] = 5
  if in_kharif and in_rabi and not(in_zaid):
    new_df['CODE'] = 4
  if not(in_kharif) and not(in_rabi) and in_zaid:
    new_df['CODE'] = 3
  if not(in_kharif) and in_rabi and not(in_zaid):
    new_df['CODE'] = 2
  if in_kharif and not(in_rabi) and not(in_zaid):
    new_df['CODE'] = 1
  if not(in_kharif) and not(in_rabi) and not(in_zaid):
    new_df['CODE'] = 0
```
